```python
#5
price = df['price'].tolist()
MaintenanceCost = df['MaintenanceCost'].tolist()
x = [idx for idx, val in enumerate(price) if val == 'high' and MaintenanceCost[idx] != 'low']
print(x)
```

```python
#7
price = df['price'].tolist()
NumberofPassengers=df['NumberofPassengers'].tolist()

a = [val for idx, val in enumerate(
    NumberofPassengers) if price[idx] == 'med']
print(a)
```

```python
#8
price = df['price'].tolist()
MaintenanceCost = df['MaintenanceCost'].tolist()

highnotLow = [i for i in range (len(price)) if price[i] == 'high' and MaintenanceCost[i] != 'low']
print(highnotLow)
```

```python
#NestedListComprehension
nlist = [[1, 2, 3], ['A', 'B', 'C'], [4, 5], ['D', 'E']]
New_list = [val for sub in nlist for val in sub]
```
