# Imputation method

```python
cars['normalized_losses'].replace('?', np.nan , inplace=True)
cars['normalized_losses'] = cars['normalized_losses'].astype("float")
cars['normalized_losses'].fillna(cars["normalized_losses"].mean(),inplace = True)
```

```python
cars['num_of_doors'].replace('?', np.nan, inplace=True)

```

```python
cars['bore'].replace('?', np.nan, inplace=True)
cars['bore'] = cars['bore'].astype("float")
cars['bore'].fillna(cars["bore"].mean(),inplace = True)
```

```python
cars['stroke'].replace('?', np.nan, inplace=True)
cars['stroke'] = cars['stroke'].astype("float")
cars['stroke'].fillna(cars["stroke"].mean(),inplace = True)
```

```python
cars['horsepower'].replace('?', np.nan, inplace=True)
cars['horsepower'] = cars['horsepower'].astype("float")
cars['horsepower'].fillna(cars["horsepower"].mean(),inplace = True)
```

```python
cars['peak-rpm'].replace('?', np.nan, inplace=True)
cars['peak-rpm'] = cars['peak-rpm'].astype("float")
cars['peak-rpm'].fillna(cars["peak-rpm"].mean(),inplace = True)
```

```python
cars['price'].replace('?', np.nan, inplace=True)
cars['price'] = cars['price'].astype("float")
cars['price'].fillna(cars["price"].mean(),inplace = True)
```
