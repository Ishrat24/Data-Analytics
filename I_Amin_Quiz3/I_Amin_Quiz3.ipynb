# Scaling

```python
df['ResidualSugar'].fillna(df["ResidualSugar"].mean(),inplace = True)
```

```python
df['Chlorides'].fillna(df["Chlorides"].mean(),inplace = True)
```

```python
df['FreeSulfurDioxide'].fillna(df["FreeSulfurDioxide"].mean(),inplace = True)
```

```python
df['TotalSulfurDioxide'].fillna(df["TotalSulfurDioxide"].mean(),inplace = True)
```

```python
df['pH'].fillna(df["pH"].mean(),inplace = True)
```

```python
df['Sulphates'].fillna(df["Sulphates"].mean(),inplace = True)
```

```python
df['Alcohol'].fillna(df["Alcohol"].mean(),inplace = True)
```

```python
df['STARS'].fillna(df["STARS"].mean(),inplace = True)
```

```python
df
```

```python
df.isnull().sum()
```

```python
# If we look at the dataframe, compare to other columns 
# ResidualSugar,FreeSulfurDioxide and TotalSulfurDioxide have large numbers
 # So i will apply scaling method those three columns
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
mx = MinMaxScaler()
```

```python
df['ResidualSugar'] = mx.fit_transform(df['ResidualSugar'].values.reshape(-1,1))
```

```python
df['FreeSulfurDioxide'] = mx.fit_transform(df['FreeSulfurDioxide'].values.reshape(-1,1))
```

```python
df['TotalSulfurDioxide'] = mx.fit_transform(df['TotalSulfurDioxide'].values.reshape(-1,1))
```

```python
plt.figure(figsize=(10,8), dpi=200)
sns.boxplot(df["TotalSulfurDioxide"])
```

```python
# There is a slight, but not insignificant difference here. 
# Demonstrating that the re-scaling of a few variables had no effect on the full data frame.
# We wanted to achieve this since replacing them with the'mean' value usually preserves data integrity.
```
